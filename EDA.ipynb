### Encoding 


```python
# Encoding de RainTomorrow
df['RainTomorrow_encoded'] = df['RainTomorrow'].map({'No': 0, 'Yes': 1})
```

```python
df.RainTomorrow_encoded.value_counts()
```

### Graficar

```python
df.RainTomorrow_encoded.value_counts().plot(kind="bar")
plt.title("Frecuencia de lluvia maÃ±ana")
plt.xlabel("Lluvia maÃ±ana")
plt.ylabel("Frecuencia")
plt.xticks(rotation=0)
```

```python
df.MaxTemp.hist()
plt.title("Histograma de temperatura mÃ¡xima")
plt.xlabel("Temperatura mÃ¡xima (Â°C)")
plt.ylabel("Frecuencia")
plt.xticks(rotation=0)
```

### ValidaciÃ³n de datos atipicos
